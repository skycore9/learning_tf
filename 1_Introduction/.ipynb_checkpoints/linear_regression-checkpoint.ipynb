```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_X = np.asarray([3.3,4.4,5.5,6.71,6.93,4.168,9.779,6.182,7.59,2.167,
                         7.042,10.791,5.313,7.997,5.654,9.27,3.1])
train_Y = np.asarray([1.7,2.76,2.09,3.19,1.694,1.573,3.366,2.596,2.53,1.221,
                         2.827,3.465,1.65,2.904,2.42,2.94,1.3])

n_samples = train_X.shape[0]
```

```python
# parameters
leanring_rate = 0.01
train_epochs = 1000
display_step = 1

# input
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# weights
weights = tf.Variable(np.random.randn())
basis = tf.Variable(np.random.randn())

# ouput
pred = tf.multiply(X, weights) + basis

# cost
loss = tf.reduce_sum(tf.pow(Y - pred, 2)) / (2*n_samples)

# optimizer
optimizer = tf.train.GradientDescentOptimizer(leanring_rate).minimize(loss)

# initializer
init = tf.global_variables_initializer()


# train
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(train_epochs):
        for x, y in zip(train_X, train_Y):
            sess.run(optimizer, feed_dict={X:x, Y:y})
            
        if (epoch + 1) % display_step == 0:
            cost = sess.run(loss, feed_dict={X:train_X, Y:train_Y})
            print("Epoch = %04d" % (epoch+1), "Cost = %6f" % cost,\
                  "Weights = %6f" % sess.run(weights), "Baises = %6f" % sess.run(basis))
    
    print("Training Done!")
    
    cost = sess.run(loss, feed_dict={X:train_X, Y:train_Y})
    print("Final Cost = %6f" % cost,\
                  "Weights = %6f" % sess.run(weights), "Baises = %6f" % sess.run(basis))
    plt.plot(train_X, train_Y, 'ro', label="Original data")
    plt.plot(train_X, sess.run(weights) * train_X + sess.run(basis), label='Fitted line' )
    plt.legend()
    plt.show()
```
