```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data

mnist = input_data.read_data_sets('/tmp/data/', one_hot=True)
```

```python
train_X, train_Y = mnist.train.next_batch(5000)
test_X, test_Y = mnist.test.next_batch(200)

tr = tf.placeholder(tf.float32, [None, 784])
te = tf.placeholder(tf.float32, [784])

# L1 distance
distance = tf.reduce_sum(tf.abs(tr - te), reduction_indices=1)

# pred
pred = tf.arg_min(distance, 0)


init = tf.global_variables_initializer()
```

```python
accuracy = 0

with tf.Session() as sess:
    sess.run(init)
    
    for idx in range(len(test_X)):
        nn_index = sess.run(pred, feed_dict={tr:train_X, te: test_X[idx,:]})
        print('Predict Class', np.argmax(train_Y[nn_index]), 'True Class', np.argmax(test_Y[idx]))
        if np.argmax(train_Y[nn_index]) == np.argmax(test_Y[idx]):
            accuracy += 1
            
    print("Training Finish!!!!")
    print('Accuracy = %f.' % (accuracy / 200.0))
```
